```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sparsities = np.linspace(0.0, 0.95, 20)

baseline_path = "results/baseline-0-0.txt"
lines = open(baseline_path, "r").readlines()
baseline = [float(x.strip()) for x in lines]
baseline = [sum(baseline) / len(baseline)] * len(sparsities)
```

## No sampling, no converison

```python
csr_csc_csc = open("results/csr-csc-csc-0-0.txt", "r").readlines()
csr_csc_csc = [float(x.strip()) for x in csr_csc_csc]

dd_csc_csc = open("results/dd-csc-csc-0-0.txt", "r").readlines()
dd_csc_csc = [float(x.strip()) for x in dd_csc_csc]

csr_dd_csc = open("results/csr-dd-csc-0-0.txt", "r").readlines()
csr_dd_csc = [float(x.strip()) for x in csr_dd_csc]

csr_csc_dd = open("results/csr-csc-dd-0-0.txt", "r").readlines()
csr_csc_dd = [float(x.strip()) for x in csr_csc_dd]

plt.plot(sparsities, csr_csc_dd, label="Q = CSR, K = CSC, V = DD")
plt.plot(sparsities, csr_csc_csc, label="Q = CSR, K = CSC, V = CSC")
plt.plot(sparsities, csr_dd_csc, label="Q = CSR, K = DD, V = CSC")
plt.plot(sparsities, dd_csc_csc, label="Q = DD, K = CSC, V = CSC")
plt.plot(sparsities, baseline, label="Q = DD, K = DD, V = DD")
plt.title("Attention with varying input matrix formats (with no sampling and no conversion)")
plt.xlabel("Sparsity ratio")
plt.ylabel("Runtime (s)")
plt.legend()
plt.show()
```
